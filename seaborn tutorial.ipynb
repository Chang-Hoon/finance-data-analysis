# Plotting with categorical data
- https://seaborn.pydata.org/tutorial/categorical.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
# sns.set_theme(style="ticks", color_codes=True)
```

## Categorical scatterplots

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.catplot(x='day', y="total_bill", data=tips)
```

```python
sns.catplot(x='day', y="total_bill", jitter=False, data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips, hue='sex', aspect=2)
```

```python
sns.catplot(x="size", y="total_bill", data=tips)
```

```python
sns.catplot(x="smoker", y="tip", order=['No', 'Yes'], data=tips)
```

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```

## Distributions of observations within categories
### Boxplots

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips, hue="smoker")
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips, hue="smoker", dodge=False)
```

```python
tips.head()
```

```python
tips["day"].unique()
```

```python
tips["weekend"] = tips["day"].isin(['Sun', 'Sat'])
tips.head()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box", dodge=False, hue="weekend")
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
sns.catplot(data=diamonds.sort_values("color"), x="color", y="price", kind="boxen")
```

### Violinplots

```python
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", data=tips)
```

```python
sns.violinplot(data=tips, x="total_bill", y="day", hue="sex",
              bw=.15, cut=0)
```

```python
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", bw=.15, cut=0, data=tips)
```

```python
sns.catplot(y="total_bill", x="day", hue="sex",
            kind="violin", split=True, data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="violin", inner="stick", split=True,
            palette="pastel", data=tips)
```

```python
g = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips, aspect=1.5)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=g.ax)
```

# Statistical estimation within categories
## Bar plots

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
g = sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
```

```python
g = sns.catplot(y="survived", hue="class", kind="count", data=titanic)
```

```python
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic)
```

```python
sns.catplot(y="deck", hue="class", kind="count",
            palette="pastel", edgecolor=".6",
            data=titanic)
```

## Point plots

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
```

```python
g = sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic)
```

# Plotting â€œwide-formâ€ data

```python
iris = sns.load_dataset("iris")
sns.catplot(data=iris, orient="h", kind="box")
```

```python
g = sns.violinplot(x=iris.species, y=iris.sepal_length)
```

```python
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c")
```

# Showing multiple relationships with facets

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.7,
            kind="swarm", data=tips)
```

```python
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log")
```
